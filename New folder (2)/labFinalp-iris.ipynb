```python
import pandas as pd
from sklearn.datasets import fetch_covtype
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
covtype = fetch_covtype()
print(covtype)
```

```python
df = pd.DataFrame(covtype.data, columns=covtype.feature_names)
df['Cover_Type'] = covtype.target
```

```python
X = covtype.data
plt.figure(figsize=(8, 4))
sns.countplot(x='Cover_Type', data=df)
plt.title("Distribution of Cover_Type")
plt.xlabel("Cover Type")
plt.ylabel("Frequency")
plt.show()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
features = ['Elevation', 'Aspect', 'Vertical_Distance_To_Hydrology', 'Horizontal_Distance_To_Roadways', 'Cover_Type']
sns.pairplot(df[features], hue='Cover_Type', corner=True)
plt.show()
```

```python
sns.pairplot(data = df,vars=['Elevation', 'Aspect', 'Slope', 'Cover_Type'], hue='Cover_Type')
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df[['Elevation', 'Aspect', 'Slope', 'Cover_Type'][:-1]].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
