```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Sports.csv')

df.head()
```

```python
df.info()
df.columns
df.describe()
```

```python
sns.scatterplot(data=df, x="Height", y="Weight", hue="Sports")
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler

x = df.drop(['Sports'],axis=1)
y = df['Sports']
```

```python
x
```

```python
y
```

```python
df.info()
```

```python
le = LabelEncoder()

y_encoded = le.fit_transform(y)
y_encoded
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

x_train, x_test, y_train, y_test = train_test_split(x,y_encoded, test_size= 0.2, random_state=42)
```

```python
x_train
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
scaller = StandardScaler()

x_train_scalled = scaller.fit_transform(x_train)
x_train_scalled
```

```python
x_test_scalled = scaller.transform(x_test)
x_test_scalled
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=3)
model.fit(x_train_scalled,y_train)
```

```python
p = model.predict(x_test_scalled)
p
```

```python
y_test
```

```python
accuracy = accuracy_score(y_test, p)
print("Accuracy:", accuracy)
```

```python
cm = confusion_matrix(y_test, p)
cm
```
